# Plot results by model type

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', 10)
```

```python
results = pd.read_csv('results/all_results.csv', sep=';')#, index_col='Model')
```

```python
results
```

```python
seq_col_brew = sns.color_palette("Greys", 8)
sns.set_palette(seq_col_brew)

subplot_no = 1
plt.figure(1)

# plot prices
plt.subplot(221)
sns.violinplot(x='Model', y='Acc.', data=results)
plt.legend()

# plot returns
plt.subplot(222)
sns.violinplot(x='Model', y='Returns', data=results)    

plt.subplots_adjust(top=1, bottom=-1, left=0, right=1.8, hspace=0.2, wspace=0.2)

plt.savefig('plots/results_violin_model_ret.png', bbox_inches='tight', transparent=True, )
```
